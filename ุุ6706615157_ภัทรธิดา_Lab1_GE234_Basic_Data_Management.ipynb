
### **4.2 การรวมข้อมูลเชิงพื้นที่และตาราง**
รวมข้อมูล Shapefile กับข้อมูลประชากรเพื่อแสดงข้อมูลประชากรในแต่ละจังหวัด โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. โหลดข้อมูลแผนที่
# แก้ไขพาธไฟล์ GeoPackage ไปยังไฟล์ที่เคยใช้งานได้สำเร็จ
geo_data = gpd.read_file('/content/drive/MyDrive/Classroom/GE234 Basic Programming/populationthailand_th.gpkg')

# 2. โหลดข้อมูลประชากร
df = pd.read_csv('/content/drive/MyDrive/Classroom/GE234 Basic Programming/Stat_pop.csv')

# --- คลีนข้อมูลเล็กน้อยเพื่อให้พร้อมใช้งาน ---
# กรองเอาเฉพาะระดับจังหวัด (รหัสจังหวัดที่ไม่ใช่ 0)
df = df[df['รหัสจังหวัด'] != '0']

# แปลงคอลัมน์ประชากรจาก String (ที่มีคอมม่า) ให้เป็นตัวเลข (Float/Int)
df['จำนวนประชากรทั้งหมด'] = df['จำนวนประชากรทั้งหมด'].str.replace(',', '').astype(float)
# ---------------------------------------

# 3. รวมข้อมูลด้วย PCODE (แม่นยำที่สุด)
merged_data = geo_data.merge(df, left_on='adm1_pcode', right_on='รหัสจังหวัด')

# 4. แสดงผล
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
merged_data.plot(column='จำนวนประชากรทั้งหมด_y', # แก้ไขชื่อคอลัมน์เป็น 'จำนวนประชากรทั้งหมด_y'
                 cmap='YlGn',
                 legend=True,
                 ax=ax,
                 legend_kwds={'label': "จำนวนประชากรทั้งหมด (คน)"})

plt.title('จำนวนประชากรในประเทศไทย', fontsize=16)
ax.set_axis_off() # ปิดเส้นแกน x, y เพื่อความสวยงาม
plt.show()
```

สำหรับการ merge พบว่าถ้าหากเราต้องการให้มันง่ายเราควรเลือกไฟล์ที่มีบางอย่างที่เหมือนกันเพื่อให้สามารถรวมกันได้ ในกรณีของหนูตอนแรกใช้จังหวัดแต่ว่ามีข้อผิดพลาดเกิดขึ้นเกี่ยวกับภาษาที่ต่างกันระว่างไทยและอังกฤษ หากยังรั้นที่จะใช้จังหวัดต่อไปการคลีนและปรับแก้จะยาวมากและยากที่จะอ่านเข้าใจ หนูเลยตัดสินใจใช้รหัสจังหวัดแทน สิ่งนี้หากมองดีๆจะพบว่าหลักการคล้ายๆกับการ join ใน ArcGIS

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. โหลดข้อมูลแผนที่
# แก้ไขพาธไฟล์ GeoPackage ไปยังไฟล์ที่เคยใช้งานได้สำเร็จ
geo_data = gpd.read_file('/content/drive/MyDrive/Classroom/GE234 Basic Programming/populationthailand_th.gpkg')

# 2. โหลดข้อมูลประชากร
df = pd.read_csv('/content/drive/MyDrive/Classroom/GE234 Basic Programming/Stat_pop.csv')

# --- คลีนข้อมูลเล็กน้อยเพื่อให้พร้อมใช้งาน ---
# กรองเอาเฉพาะระดับจังหวัด (รหัสจังหวัดที่ไม่ใช่ 0)
df = df[df['รหัสจังหวัด'] != '0']

# แปลงคอลัมน์ประชากรจาก String (ที่มีคอมม่า) ให้เป็นตัวเลข (Float/Int)
df['จำนวนประชากรทั้งหมด'] = df['จำนวนประชากรทั้งหมด'].str.replace(',', '').astype(float)
df['จำนวนประชากรชาย'] = df['จำนวนประชากรชาย'].str.replace(',', '').astype(float) # เพิ่มการแปลงคอลัมน์ประชากรชาย
# ---------------------------------------

# 3. รวมข้อมูลด้วย PCODE (แม่นยำที่สุด)
merged_data = geo_data.merge(df, left_on='adm1_pcode', right_on='รหัสจังหวัด')

# 4. แสดงผล
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
merged_data.plot(column='จำนวนประชากรชาย_y', # เปลี่ยนเป็น 'จำนวนประชากรชาย_y'
                 cmap='YlGnBu',
                 legend=True,
                 ax=ax,
                 scheme='NaturalBreaks', # เพิ่ม classification scheme
                 k=5, # จำนวนคลาสที่ต้องการ
                 legend_kwds={'title': "จำนวนประชากรชาย (คน)"}) # แก้ไขชื่อ legend

plt.title('จำนวนประชากรชายในประเทศไทย', fontsize=16)
ax.set_axis_off() # ปิดเส้นแกน x, y เพื่อความสวยงาม
plt.show()
```
